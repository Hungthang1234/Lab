```python
import requests
from bs4 import BeautifulSoup

# Bước 1: Lấy danh sách link bài viết từ trang chủ
url = "https://vnexpress.net/khoa-hoc-cong-nghe"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Tìm tất cả các thẻ tiêu đề bài viết
titles = soup.find_all("h3", class_="title-news")  # Có thể là h3 thay vì h2
links = [title.find("a")["href"] for title in titles if title.find("a")]

# Bước 2: Duyệt từng link bài viết để lấy chi tiết
for link in links:
    news = requests.get(link)
    soup = BeautifulSoup(news.content, "html.parser")

    title_tag = soup.find("h1", class_="title-news")
    abstract_tag = soup.find("p", class_="description")

    title = title_tag.text.strip() if title_tag else "Không có tiêu đề"
    abstract = abstract_tag.text.strip() if abstract_tag else "Không có mô tả"

    print('Tiêu đề: ' + title)
    print('Mô tả : ' + abstract)
    print('______________________________________________')
```
