```python
import requests
from bs4 import BeautifulSoup

# Bc 1: Ly danh sch link bi vit t trang ch
url = "https://vnexpress.net/khoa-hoc-cong-nghe"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Tm tt c cc th tiu  bi vit
titles = soup.find_all("h3", class_="title-news")  # C th l h3 thay v h2
links = [title.find("a")["href"] for title in titles if title.find("a")]

# Bc 2: Duyt tng link bi vit  ly chi tit
for link in links:
    news = requests.get(link)
    soup = BeautifulSoup(news.content, "html.parser")

    title_tag = soup.find("h1", class_="title-news")
    abstract_tag = soup.find("p", class_="description")

    title = title_tag.text.strip() if title_tag else "Khng c tiu "
    abstract = abstract_tag.text.strip() if abstract_tag else "Khng c m t"

    print('Tiu : ' + title)
    print('M t : ' + abstract)
    print('______________________________________________')
```
