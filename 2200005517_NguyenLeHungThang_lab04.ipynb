```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Tải dữ liệu
data = load_iris()
X = data.data
y = data.target
# Chia dữ liệu thành tập huấn luyện và kiểm thử
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Khởi tạo mô hình với k=3
knn = KNeighborsClassifier(n_neighbors=3)
# Huấn luyện mô hình
knn.fit(X_train, y_train)
# Dự đoán và đánh giá
y_pred = knn.predict(X_test)
print("Độ chính xác của KNN:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
# Khởi tạo mô hình
nb = GaussianNB()
# Huấn luyện mô hình
nb.fit(X_train, y_train)
# Dự đoán và đánh giá
y_pred = nb.predict(X_test)
print("Độ chính xác của Naive Bayes:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
# Khởi tạo mô hình
tree = DecisionTreeClassifier(random_state=42)
# Huấn luyện mô hình
tree.fit(X_train, y_train)
# Dự đoán và đánh giá
y_pred = tree.predict(X_test)
print("Độ chính xác của Decision Tree:", accuracy_score(y_test, y_pred))
```

```python
print("Độ chính xác của các mô hình:")
print("KNN:", accuracy_score(y_test, knn.predict(X_test)))
print("Naive Bayes:", accuracy_score(y_test, nb.predict(X_test)))
print("Decision Tree:", accuracy_score(y_test, tree.predict(X_test)))
```
