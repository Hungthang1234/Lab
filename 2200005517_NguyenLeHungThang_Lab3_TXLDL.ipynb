```python
df = pd.read_csv('/content/drive/MyDrive/data_test_lab3.csv')
df.head()
```

```python
# 2.2. Tìm trong DataFrame những dữ liệu bị thiếu
missing_data = df.isnull()
print("Các giá trị bị thiếu:\n", missing_data)

# 2.3. Xem cột NUM_BEDROOMS và tìm những giá trị NaN
nan_bedrooms = df[df["NUM_BEDROOMS"].isnull()]
print("\nCác hàng có NUM_BEDROOMS là NaN:\n", nan_bedrooms)

# 2.4. Tổng giá trị còn thiếu trong toàn bộ DataFrame
total_missing = df.isnull().sum().sum()
print("\nTổng số giá trị bị thiếu trong toàn bộ DataFrame:", total_missing)

# 2.5. Điền giá trị 126 vào các ô bị thiếu trong cột “ST_NUM”
df["ST_NUM"].fillna(126, inplace=True)

```

```python
# Chuyển cột NUM_BEDROOMS về kiểu số, các giá trị không chuyển được sẽ thành NaN
df["NUM_BEDROOMS"] = pd.to_numeric(df["NUM_BEDROOMS"], errors="coerce")

# Sau đó tính median và điền vào các ô thiếu
median_bedrooms = df["NUM_BEDROOMS"].median()
df["NUM_BEDROOMS"].fillna(median_bedrooms, inplace=True)

# In kết quả kiểm tra
print("Giá trị trung vị của NUM_BEDROOMS là:", median_bedrooms)
print(df["NUM_BEDROOMS"])

```
