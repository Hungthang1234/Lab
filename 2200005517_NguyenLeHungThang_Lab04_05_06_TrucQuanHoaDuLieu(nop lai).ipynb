```python
# Ti d liu tips c sn trong seaborn
tips = sns.load_dataset('tips')

# Xem cc ct trong d liu
print("Cc ct trong d liu:")
print(tips.columns)

# Tnh tng tin tip theo tng ngy
tip_per_day = tips.groupby('day')['tip'].sum()

# V biu  ct (Bar Chart)
plt.figure(figsize=(8,6))
tip_per_day.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Hin th s tin thng trong ngy ($)', fontsize=14)
plt.xlabel('Ngy trong tun', fontsize=12)
plt.ylabel('Tip (Tin thng)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
cars = ['AUDI', 'BMW','FORD','TOYOTA ', 'KIA','MERCERDES']
data1 = [23,17,45,39,10,31]
explode =(0.1,0,0.1,0,0,0)

colors = ['orange', 'cyan', 'brown', 'green', 'yellow', 'beige']
wp = {'linewidth' : 1 ,'edgecolor' : "green"}

def func(pct,allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} g)".format(pct,absolute)


fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(
    data1,
    autopct=lambda pct: func(pct, data1),
    explode=explode,
    labels=cars,
    shadow=True,
    colors=colors,
    startangle=90,
    wedgeprops=wp,
    textprops=dict(size=12, color='black')  # 'colors' -> 'color'
)

ax.legend(wedges, cars,
          title="Cars",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
plt.show()
```
