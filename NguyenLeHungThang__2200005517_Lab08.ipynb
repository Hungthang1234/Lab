```python
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.impute import KNNImputer
```

```python
# ƒê·ªçc d·ªØ li·ªáu t·ª´ URL (ƒë√¢y l√† d·ªØ li·ªáu Pima Indians Diabetes)
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(url, names=names)

# T√°ch d·ªØ li·ªáu th√†nh X (bi·∫øn ƒë·∫ßu v√†o) v√† y (nh√£n)
X, y = data.values[:, :-1], data.values[:, -1]

# Chia d·ªØ li·ªáu th√†nh t·∫≠p hu·∫•n luy·ªán v√† ki·ªÉm tra (80/20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)

# In k√≠ch th∆∞·ªõc t·∫≠p hu·∫•n luy·ªán v√† ki·ªÉm tra
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
```

```python

# ƒê·ªçc d·ªØ li·ªáu t·ª´ file csv online
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(url, names=names)

# T√°ch X (bi·∫øn ƒë·∫ßu v√†o), kh√¥ng l·∫•y c·ªôt 'class'
X = data.drop('class', axis=1)

# T·∫°o b·ªô l·ªçc VarianceThreshold
selector = VarianceThreshold()  # Gi·ªØ l·∫°i c√°c c·ªôt c√≥ ph∆∞∆°ng sai l·ªõn h∆°n 0

# √Åp d·ª•ng l·ªçc ph∆∞∆°ng sai th·∫•p
X_sel = selector.fit_transform(X)

# In k√≠ch th∆∞·ªõc d·ªØ li·ªáu tr∆∞·ªõc v√† sau l·ªçc
print("Tr∆∞·ªõc khi l·ªçc:", X.shape)
print("Sau khi l·ªçc:", X_sel.shape)
```

```python
# ƒê·ªçc d·ªØ li·ªáu
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(url, names=names)

# ƒê√°nh d·∫•u c√°c gi√° tr·ªã 0 ·ªü c√°c c·ªôt y t·∫ø l√† missing (0 kh√¥ng h·ª£p l√Ω ·ªü c√°c bi·∫øn n√†y)
columns_to_mark = ['plas', 'pres', 'skin', 'test', 'mass']
data[columns_to_mark] = data[columns_to_mark].replace(0, np.nan)

# In s·ªë l∆∞·ª£ng missing values m·ªói c·ªôt
print("S·ªë l∆∞·ª£ng gi√° tr·ªã thi·∫øu:")
print(data.isnull().sum())

# X√≥a d√≤ng ch·ª©a b·∫•t k·ª≥ gi√° tr·ªã thi·∫øu n√†o
data_dropna = data.dropna()

# T√°ch X v√† y
X = data_dropna.drop('class', axis=1)
y = data_dropna['class']

# Chia train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)

# Hu·∫•n luy·ªán m√¥ h√¨nh Logistic Regression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# D·ª± ƒëo√°n v√† t√≠nh accuracy
y_pred = model.predict(X_test)
print("ƒê·ªô ch√≠nh x√°c sau khi x√≥a d·ªØ li·ªáu thi·∫øu:", accuracy_score(y_test, y_pred))
```

```python
# ƒê·ªçc d·ªØ li·ªáu
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(url, names=names)

# ƒê√°nh d·∫•u gi√° tr·ªã 0 l√† missing (·ªü c√°c c·ªôt y t·∫ø)
columns_to_mark = ['plas', 'pres', 'skin', 'test', 'mass']
data[columns_to_mark] = data[columns_to_mark].replace(0, np.nan)

# T√°ch X v√† y
X = data.drop('class', axis=1)
y = data['class']

# Thay th·∫ø missing b·∫±ng trung b√¨nh c·ªôt
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Chia d·ªØ li·ªáu
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=1)

# Hu·∫•n luy·ªán m√¥ h√¨nh
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# D·ª± ƒëo√°n v√† ƒë√°nh gi√°
y_pred = model.predict(X_test)
print("ƒê·ªô ch√≠nh x√°c sau khi thay th·∫ø b·∫±ng trung b√¨nh:", accuracy_score(y_test, y_pred))
```

```python

# ƒê·ªçc d·ªØ li·ªáu
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(url, names=names)

# ƒê√°nh d·∫•u gi√° tr·ªã thi·∫øu
columns_to_mark = ['plas', 'pres', 'skin', 'test', 'mass']
data[columns_to_mark] = data[columns_to_mark].replace(0, np.nan)

# T√°ch X v√† y
X = data.drop('class', axis=1)
y = data['class']

# Thay th·∫ø missing b·∫±ng KNN (m·∫∑c ƒë·ªãnh k=5)
imputer = KNNImputer(n_neighbors=5)
X_imputed = imputer.fit_transform(X)

# Chia t·∫≠p d·ªØ li·ªáu
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=1)

# Hu·∫•n luy·ªán m√¥ h√¨nh
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# D·ª± ƒëo√°n v√† ƒë√°nh gi√°
y_pred = model.predict(X_test)
print("ƒê·ªô ch√≠nh x√°c sau khi d√πng KNNImputer:", accuracy_score(y_test, y_pred))
```

## üìå T·ªïng k·∫øt

Trong b√†i lab n√†y, ƒë√£ th·ª±c hi·ªán c√°c k·ªπ thu·∫≠t x·ª≠ l√Ω v√† chu·∫©n b·ªã d·ªØ li·ªáu quan tr·ªçng ƒë·ªÉ ph·ª•c v·ª• cho vi·ªác hu·∫•n luy·ªán m√¥ h√¨nh h·ªçc m√°y, d·ª±a theo s√°ch *"Data Preparation for Machine Learning"* . C·ª• th·ªÉ:

- **Ch∆∞∆°ng 4**: H·ªçc c√°ch chia d·ªØ li·ªáu ƒë√∫ng c√°ch ƒë·ªÉ tr√°nh hi·ªán t∆∞·ª£ng r√≤ r·ªâ d·ªØ li·ªáu (data leakage) b·∫±ng `train_test_split` v√† cross-validation.
- **Ch∆∞∆°ng 5**: X√°c ƒë·ªãnh v√† lo·∫°i b·ªè c√°c thu·ªôc t√≠nh kh√¥ng ph√π h·ª£p nh∆∞ c·ªôt c√≥ gi√° tr·ªã l·∫∑p, ph∆∞∆°ng sai th·∫•p, ho·∫∑c tr√πng l·∫∑p.
- **Ch∆∞∆°ng 6**: Ph√°t hi·ªán v√† x·ª≠ l√Ω c√°c gi√° tr·ªã ngo·∫°i lai b·∫±ng c√°c ph∆∞∆°ng ph√°p nh∆∞ ƒë·ªô l·ªách chu·∫©n (standard deviation) v√† kho·∫£ng t·ª© ph√¢n v·ªã (IQR).
- **Ch∆∞∆°ng 7**: ƒê√°nh d·∫•u v√† x√≥a d·ªØ li·ªáu thi·∫øu ƒë·ªÉ ƒë·∫£m b·∫£o t√≠nh ƒë·∫ßy ƒë·ªß v√† ƒë·ªô tin c·∫≠y c·ªßa t·∫≠p d·ªØ li·ªáu.
- **Ch∆∞∆°ng 8**: Thay th·∫ø d·ªØ li·ªáu thi·∫øu b·∫±ng gi√° tr·ªã th·ªëng k√™ (mean, median) b·∫±ng c√°ch s·ª≠ d·ª•ng `SimpleImputer`.
- **Ch∆∞∆°ng 9**: Thay th·∫ø d·ªØ li·ªáu thi·∫øu b·∫±ng ph∆∞∆°ng ph√°p KNN (k-nearest neighbors) s·ª≠ d·ª•ng `KNNImputer`, gi√∫p m√¥ ph·ªèng t·ªët h∆°n m·ªëi quan h·ªá gi·ªØa c√°c bi·∫øn.

K·∫øt qu·∫£ m√¥ h√¨nh ƒë∆∞·ª£c ƒë√°nh gi√° b·∫±ng ƒë·ªô ch√≠nh x√°c (accuracy) tr∆∞·ªõc v√† sau khi x·ª≠ l√Ω d·ªØ li·ªáu, qua ƒë√≥ cho th·∫•y vi·ªác chu·∫©n b·ªã d·ªØ li·ªáu c√≥ ·∫£nh h∆∞·ªüng r·∫•t l·ªõn ƒë·∫øn ch·∫•t l∆∞·ª£ng m√¥ h√¨nh h·ªçc m√°y.
