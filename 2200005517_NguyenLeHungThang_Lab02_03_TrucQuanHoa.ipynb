```python
import matplotlib.pyplot as plt

# Dữ liệu cho 2 đường
x1 = [2,3,8,15]
x2 = [2,3,6,9,15]
y1 = [10,20,15,55]  # Đường 1: y = x^2
y2 = [10,10,20,15,45]  # Đường 2: y = -x^2

# Vẽ 2 đường trên cùng một biểu đồ
plt.plot(x1, y1, label='1', color='r', linestyle='-', marker='o')  # Đường 1
plt.plot(x2, y2, label='2', color='g', linestyle='--', marker='o')  # Đường 2

# Thêm tiêu đề và nhãn cho các trục
plt.title('Vẽ 2 đường trên biểu đồ')
plt.xlabel('X')
plt.ylabel('Y')

# Thêm chú thích
plt.legend()

# Hiển thị biểu đồ
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dữ liệu tọa độ
x = [20, 22, 24, 26, 28, 30]


y1 = [5, 3, 5, 4, 5, 5]
y2 = [4,4.5,5,3,4,7]
y3 = [5, 4, 6, 4, 6, 8]

# Vẽ các đường đồ thị với các màu sắc và chú thích khác nhau
plt.plot(x, y1, label='Xang', color='blue' )  # Đường màu xanh dương
plt.plot(x, y2, label='Dau', color='green')  # Đường màu xanh lá
plt.plot(x, y3, label='Nhot', color='red')  # Đường màu đỏ

# Thêm tiêu đề và nhãn cho các trục
plt.title('Biểu đồ giá nhiên liệu năm 2020')
plt.xlabel('Giá nhiên liệu VND')
plt.ylabel('Tháng')

# Thêm chú thích
plt.legend()

# Hiển thị lưới
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```

```python


# Tạo các giá trị x từ -20 đến 20 với 80 điểm
x = np.linspace(-20, 20, 80)

# Tính giá trị sin của x
y = np.sin(x)

# Vẽ biểu đồ
plt.plot(x, y, color='r')

# Thêm tiêu đề và nhãn cho các trục
plt.title('Biểu đồ hàm sin')
plt.xlabel('x')
plt.ylabel('y = sin(x)')

# Thêm chú thích
plt.legend()

# Hiển thị lưới
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```

```python
x1 = [0, 1, 2, 3, 4]
y1 = [1, 2, 3, 4, 10]

x2 = [0, 1, 2, 3, 4]
y2 = [10, 4, 3, 2, 1]

x3 = [2.5, 2.5, 2.5, 1.5, 0.5]
y3 = [1, 3, 5, 7, 10]

x4 = [0, 1.4, 2, 5, 6]
y4 = [3, 2, 3, 5, 8]

# Vẽ các đường đồ thị
plt.plot(x1, y1, label='Đường 1', color='g',  marker='o')
plt.plot(x2, y2, label='Đường 2', color='r', marker='o')
plt.plot(x3, y3, label='Đường 3', color='b', marker='o')
plt.plot(x4, y4, label='Đường 4', color='y', marker='o')

# Thêm tiêu đề và nhãn cho các trục
plt.title('Biểu đồ các đường Line')
plt.xlabel('X')
plt.ylabel('Y')

# Thêm chú thích
plt.legend()

# Hiển thị lưới
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```

```python
# Dữ liệu cho các điểm
x1 = [0, 1, 4, 6, 5]
y1 = [1, 2, 3, 5, 6]

x2 = [0, 1, 2, 7]
y2 = [10, 4, 3, 2]

x3 = [2.5, 2.5, 2.5, 1.5, 1 ]
y3 = [1, 3, 5, 7, 10]

# Tạo 3 trục con (3 biểu đồ)
fig, (ax1 , ax2 ,ax3) = plt.subplots(1,3, figsize=(14, 7), sharey = True , dpi = 120)
# Biểu đồ 1
ax1.plot(x1, y1, label='Đường 1', color='blue', marker='o')
ax1.set_title('Biểu đồ 1')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.grid(True)
ax1.legend()

# Biểu đồ 2
ax2.plot(x2, y2, label='Đường 2', color='green', marker='o')
ax2.set_title('Biểu đồ 2')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.grid(True)
ax2.legend()

# Biểu đồ 3
ax3.plot(x3, y3, label='Đường 3', color='red',  marker='o')
ax3.set_title('Biểu đồ 3')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.grid(True)
ax3.legend()



# Hiển thị biểu đồ
plt.show()
```
