```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

# c php cho php chiu 3-D
ax = plt.axes (projection ='3d')
# xc nh c 3 trc
z= np.linspace(0, 1, 100)
x= z * np.sin(25* z)
y= z * np.cos(25* z)
c = x + y
ax.scatter3D(x, y, z, c=c)
# plotting
ax.set_title('30 line plot')
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

# Xc nh b mt v trc
x = np.outer(np.linspace(-2, 2, 10), np.ones(10))
y = x.copy().T
z = np.cos(x ** 2 + y ** 3)

fig = plt.figure()

# C php cho 3-0 plotting
ax = plt.axes(projection = '3d')
# plotting
ax.plot_surface(x, y, z, cmap ='viridis', edgecolor ='green')
ax.set_title('Surface plot geeks for geeks')
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

# Hm ca trc z
def f(x, y):
 return np.sin(np.sqrt(x ** 2 + y ** 2))

# Trc x v y
x = np.linspace(-1, 5, 10)
y = np.linspace(-1, 5, 10)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes (projection = '3d')
ax.plot_wireframe (X, Y, Z, color='red')
ax.set_title('wireframe');
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Xc nh cc tham s ca Mbius strip
R = 2

# nh ngha b mt Mbius strip
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(-1, 1, 100)
u, v = np.meshgrid(u, v)
x = (R+ v* np.cos(u/2)) * np.cos(u)
y = (R+ v* np.cos(u/2)) * np.sin(u)
z = v * np.sin(u/2)

# To plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#B mt ca Plot the Mobius strip
ax.plot_surface(x, y, z, alpha=0.5)

#Set thuc tnh plot
ax.set_xlabel('x')
ax.set_ylabel('Y')
ax.set_zlabel('2')
ax.set_title('Mbius Strip')

# Set gii hn plot
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])
#Show plot
plt.show()
```
