```python
df = pd.read_csv('/content/GDP 2019 2020 - Lab 7.csv')  # thay 'data.csv' bằng tên file bạn đã upload
df.head()
df.set_index('Country Name', inplace=True)
df
```

```python
n_countries = len(df)
intervals = [i*2 for i in range(75)]

wdf = df[['2019', '2020']].sort_values('2020')
wdf['2020'].plot(figsize=(15,5))

plt.xticks(intervals, wdf.iloc[intervals].index, rotation=90)
plt.show()

wdf['2020'].plot.box(vert=False, figsize=(15,1))
plt.show()
```

```python
n_countries = len(df)
intervals = [i*2 for i in range(75)]

wdf = df[['2019', '2020']].sort_values('2020')
wdf['2020'].plot(figsize=(15,5), logy=True)

plt.xticks(intervals, wdf.iloc[intervals].index, rotation=90)
plt.show()

wdf['2020'].plot.box(vert=False, figsize=(15,1), logx=True)
plt.show()
```

```python
df['log_2020'] = np.log(df['2020'])
df.log_2020.sort_values().plot()

plt.xticks(rotation=90)
plt.show()
```

```python
tdf = pd.read_csv('/content/Noise_data.csv')  # thay 'data.csv' bằng tên file bạn đã upload
tdf.head()
```

```python
tdf.drop(columns='t', inplace=True)
tdf.Signal.plot(figsize=(15,5))
plt.show()
```

```python
from statsmodels.nonparametric.kernel_regression import KernelReg
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 50, 50)
y = tdf.Signal.iloc[:50]
plt.plot(x, y, '+')

kr = KernelReg(y, x, 'c')
y_pred, y_std = kr.fit(x)

plt.plot(x, y_pred)
plt.show()
```
