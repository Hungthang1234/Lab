```python
# importing pandas as pd
import pandas as pd

# importing numpy as np
import numpy as np

# dictionary of lists
dict = {
    'Điểm dòng 1': [100, 90, np.nan, 95],
    'Điểm dòng 2': [30, 45, 56, np.nan],
    'Điểm dòng 3': [np.nan, 40, 80, 98]
}

# Tạo một dataframe từ dictionary
df = pd.DataFrame(dict)

# Hiển thị DataFrame ban đầu
print("DataFrame ban đầu:")
print(df)

# Thay thế các giá trị còn thiếu (NaN) bằng 0 với fillna()
print("\nDataFrame sau khi thay thế NaN bằng 0:")
print(df.fillna(0))
```

```python
import pandas as pd
import numpy as np

# Tạo DataFrame với các giá trị NaN
dict = {
    'Điểm dòng 1': [100, 90, np.nan, 95],
    'Điểm dòng 2': [30, 45, 56, np.nan],
    'Điểm dòng 3': [np.nan, 40, 80, 98]
}

df = pd.DataFrame(dict)

# Điền các giá trị NaN bằng giá trị liền trước đó (forward fill)
print(df.fillna(method='pad'))
```

```python
import pandas as pd
import numpy as np

# Giả sử đã có file employees.csv
data = pd.read_csv("employees.csv")

# Thay thế tất cả giá trị NaN bằng -88
data_new = data.replace(to_replace=np.nan, value=-88)

print(data_new)
```

```python
import pandas as pd

df = pd.DataFrame({
    "A": [12, 4, 5, None, 1],
    "B": [None, 2, 54, 3, None],
    "C": [20, 16, None, 3, 8],
    "D": [14, 3, None, None, 6]
})

print(df)
print("=============================")

# Nội suy giá trị NaN theo phương pháp tuyến tính
data_new = df.interpolate(method='linear', limit_direction='forward')

print(data_new)
```

```python
data = pd.read_csv("employees.csv")

new_data = data.dropna(axis=0, how='any')

print("Độ dài khung dữ liệu cũ:", len(data))
print("Độ dài khung dữ liệu mới:", len(new_data))
print("Những dòng có ít nhất 1 giá trị NaN:", len(data) - len(new_data))
```

```python
import seaborn as sns
import pandas as pd

# Load dataset Titanic từ seaborn
data = sns.load_dataset("titanic")

# Xem trước dữ liệu
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Biểu đồ thể hiện các ô có giá trị thiếu
plt.figure(figsize=(10,6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Biểu đồ thể hiện giá trị thiếu trong Titanic Dataset")
plt.show()
```

```python
# Xóa dòng có ít nhất 1 giá trị NaN
data_cleaned = data.dropna()
print(data_cleaned)

# Xóa cột có ít nhất 1 giá trị NaN
data_cleaned_col = data.dropna(axis=1)
print(data_cleaned_col)
```
